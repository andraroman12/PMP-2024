```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta


observatii_1 = ["s","s", "b", "b","s", "s", "s", "s", "b", "s"]


num_b_1 = observatii_1.count("b")
num_s_1 = observatii_1.count("s")


alpha_prior = 1
beta_prior = 1

#distributia posteriori pt primu experiment
alpha_posterior_1 = alpha_prior + num_b_1
beta_posterior_1 = beta_prior + num_s_1

observatii_2 = ["b", "b", "s", "b", "s", "s", "b", "s","s","b"]

num_b_2 = observatii_2.count("b")
num_s_2 = observatii_2.count("s")

#subpct a beta uniforma
alpha_posterior_2a = alpha_prior + num_b_2
beta_posterior_2a = beta_prior + num_s_2

#subpt b distributia a priori este distributia obtinuta in exp precendent
alpha_prior_b = alpha_posterior_1
beta_prior_b = beta_posterior_1

alpha_posterior_2b = alpha_prior_b + num_b_2
beta_posterior_2b = beta_prior_b + num_s_2


x = np.linspace(0, 1, 1000)

#pt primul exp -distributia
posterior_1_pdf = beta.pdf(x, alpha_posterior_1, beta_posterior_1)

#pt al doilea exp-distributia
posterior_2a_pdf = beta.pdf(x, alpha_posterior_2a, beta_posterior_2a)
posterior_2b_pdf = beta.pdf(x, alpha_posterior_2b, beta_posterior_2b)

#graful pt primul exp
plt.figure(figsize=(12, 6))
plt.plot(x, posterior_1_pdf, label=f"Posterior Experiment 1: Beta({alpha_posterior_1}, {beta_posterior_1})", color="blue")
plt.title("Distributia a posteriori dupa primul experiment")
plt.xlabel("Probabilitatea de a obtine 'ban'")
plt.ylabel("Densitate de probabilitate")
plt.legend()
plt.grid(True)
plt.show()

# graful pt al doilea exp
plt.figure(figsize=(12, 6))
plt.plot(x, posterior_2a_pdf, label=f"Posterior Caz a: Beta({alpha_posterior_2a}, {beta_posterior_2a})", color="green")
plt.plot(x, posterior_2b_pdf, label=f"Posterior Caz b: Beta({alpha_posterior_2b}, {beta_posterior_2b})", color="red")
plt.title("Distributia a posteriori pentru al doilea experiment")
plt.xlabel("Probabilitatea de a obtine 'ban'")
plt.ylabel("Densitate de probabilitate")
plt.legend()
plt.grid(True)
plt.show()
```
